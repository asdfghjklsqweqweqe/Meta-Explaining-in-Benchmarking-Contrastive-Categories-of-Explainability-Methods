```python
class VisualUserInterface(param.Parameterized):
    
    manipulcation_form = {
        'GC': ['Graph Name', 'Graph Index'],
        'NC': ['Node Name', 'Node Index']
    }


    ex_method_dict = {
        'GC': ['SA Graph', 'GuidedBP Graph'],
        'NC': ['SA Node', 'GuidedBP Node']
    }
    Task = param.Selector(default='GC', objects=['GC', 'NC'])
    
    selection_way = param.Selector(default='Graph Index', objects=manipulcation_form['GC'])

    gnn_model = param.Selector(default='GCN + GAP', objects=["GCN + GAP", "GCNND", "DGCNN", "DIFFPOOL", "DIFFPOOLD", "GraphSig + DNN"])

    explainable_ex_method = param.Selector(default="SA Graph", objects= ex_method_dict['GC'])

    #button = pn.widgets.Button(name='Click me', button_type='primary')
    #text = pn.widgets.TextInput(value='Ready')
    #button.clicks
    #text = pn.widgets.TextInput(value='Ready')
    random_button = pn.widgets.Button(name='Set Config and Click me', button_type='primary')
    display(random_button)
    

    
    @pn.depends('Task', watch=True)
    def _update_selection_way(self):
      selections_way = self.manipulcation_form[self.Task]
      self.param.selection_way.objects = selections_way
      self.selection_way = selections_way[0]

    @pn.depends('Task', watch=True)
    def _update_ex_method(self):
      explainable_ex_method = self.ex_method_dict[self.Task]
      self.param.explainable_ex_method.objects = explainable_ex_method
      self.explainable_ex_method = explainable_ex_method[0]


    def update_random(event):
      val = np.random.rand()
      print(val)

    def button_pressed(event):
      VisualUserInterface.update_random(event)
    
    random_button.on_click(button_pressed)


pn.panel(VisualUserInterface())
```

```python
import numpy as np
import ipywidgets as widgets

# create an output widget
rand_num_output = widgets.Output()
output = widgets.Output()

class VisualUserInterface(object):
  def __init__(self):
    self.start_button = widgets.Button(description='Click to Start...', button_style='info', layout = widgets.Layout(width='220px'))
    self.start_button.on_click(self.button_pressed)

    self.Task = widgets.Dropdown(options=["Task to Choose...", "GC", "NC"], value="Task to Choose...", description='',disabled=False, layout = widgets.Layout(width='220px'))
    self.Task.default_value = "Task to Choose..."
    self.Task.observe(self.on_change)

    self.query_way = widgets.Dropdown(options=["Please choose the task first..."], value="Please choose the task first...", description='',disabled=False, layout = widgets.Layout(width='220px'))
    self.query_way.default_value = "Please choose the task first..."

    self.gnn_model = widgets.Dropdown(options=["Model Config.", "GCN + GAP", "GCNND", "DGCNN", "DIFFPOOL", "DIFFPOOLD", "GraphSig + DNN"], value="Model Config.", description='',disabled=False, layout = widgets.Layout(width='220px'))
    self.gnn_model.default_value = "Model Config."
    
    self.explainable_method = widgets.Dropdown(options= ["Please choose the task first..."], value="Please choose the task first...", description='',disabled=False, layout = widgets.Layout(width='220px'))
    self.explainable_method.default_value = "Please choose the task first..."

    #self.input_widgets = widgets.HBox([self.Task, self.query_way, self.gnn_model, self.explainable_method])
    self.my_output_dict = {"Task":[],"Index_OR_Name":[],"Model":[], "Explainable Method":[]}

    self.RE_start_button = widgets.Button(description='click to set default values', button_style='danger', layout = widgets.Layout(width='220px'))
    self.RE_start_button.on_click(self.reset_on_click)

    self.defaulting_widgets = [self.Task, self.query_way, self.gnn_model, self.explainable_method]

  def reset_on_click(self,button):
    for widget in self.defaulting_widgets:
      widget.value = widget.default_value
  
  
  def update_random(self):
      # clear the output on every click of randomize self.val
      rand_num_output.clear_output()
      
      # execute function so it gets captured in output widget view
      with rand_num_output:
          self.val = np.random.rand()
          print(self.val)

  def button_pressed(self,*args):
    self.update_random()
    if self.Task.value != "Task to Choose..." and self.query_way.value != "Please choose the task first..." and self.explainable_method.value != "Please choose the task first...":
      self.my_output_dict["Task"].append(self.Task.value)
      self.my_output_dict["Index_OR_Name"].append(self.query_way.value)
      self.my_output_dict["Model"].append(self.gnn_model.value)
      self.my_output_dict["Explainable Method"].append(self.explainable_method.value)
      self.call_me_to_go_for_explainable_method()
      #print(self.my_output_dict)
    else:
      rand_num_output.clear_output()
      output.clear_output()


  def on_change(self,change):
    if change['type'] == 'change' and change['name'] == 'value':
      print("changed to %s" % change['new'])
      if change['new'] == "GC":
        self.query_way.options = ["Graph Name", 'Graph Index']
        self.explainable_method.options = ['SA', 'Guided BP', 'CAM', 'Grad-CAM', 'SubgraphX', 'LRP', 'Excitation BP']
      elif change['new'] == "NC":
        self.query_way.options = ["Node Name", 'Node Index']
        self.explainable_method.options = ['SA', 'Guided BP', 'CAM', 'Grad-CAM', 'SubgraphX', 'LRP', 'Excitation BP', 'GraphLime']
      else:
        self.query_way.options = ["Please choose the task first..."]
        self.explainable_method.options = ["Please choose the task first..."]

  def call_me_to_go_for_explainable_method(self):
    print(self.my_output_dict)

ClassObject = VisualUserInterface()
# display(buttonObject.random_button)  # display works but is not required if on the last line in Jupyter cell.
Box_Automation = [ClassObject.Task, output, ClassObject.query_way, ClassObject.gnn_model, ClassObject.explainable_method, ClassObject.start_button, rand_num_output, ClassObject.RE_start_button, output]
widgets.VBox(Box_Automation)
```

```python
import pandas as pd
import datetime
from datetime import *
import ipywidgets as widgets
from ipywidgets import *

############################################## The following section is usually in a seperate cell so I can 
df = {                                       # refresh my portfolio every day, but still add to the DF throughout the day
    'Datetime' : [],
    'Expire' : [],
    'Type' : [],
    'Quantity' : [],
    'Strike' : [],
    'Spot' : []
}

df = pd.DataFrame(df)
df.to_csv("portfolio.csv", index=False)
##############################################


Type = widgets.Dropdown(
    options = ['Call', 'Put'],
    value = 'Call',
    description= 'Select Binary Type',
    disabled=False,
    layout={'width': 'max-content'},
    style={'description_width': 'max-content'}
)

Quantity = widgets.BoundedIntText(value=1,
    min=1,
    max=10,
    step=1,
    description='Quantity:',
    disabled=False,
    layout={'width': 'max-content'},
    style={'description_width': 'max-content'}
)

Strike = widgets.BoundedIntText(
    min=1500,
    max=3500,
    step=1,
    description='Strike:',
    disabled=False,
    layout={'width': 'max-content'},
    style={'description_width': 'max-content'}
)

Spot = widgets.BoundedIntText(
    min=1500,
    max=3500,
    step=1,
    description='Spot:',
    disabled=False,
    layout={'width': 'max-content'},
    style={'description_width': 'max-content'}    
)


Add = widgets.Button(description="Add Data")
out = Output()

def add_on_click(b):

    dt = datetime.now()
    option = Type.value
    quant = Quantity.value
    strike = Strike.value
    spot = Spot.value

    df = pd.read_csv("portfolio.csv")
    now = datetime.now()

    add = {
        'Datetime' : dt,
        'Expire' : datetime(now.year, now.month, now.day, 14, 15,0,1),
        'Type' : option,
        'Quantity': quant,
        'Strike' : strike,
        'Spot': spot
    }

    add = pd.DataFrame(add, index=[0])

    df = pd.concat([df, add],sort=True) #ignore_index=True)
    df.to_csv("portfolio.csv", index=False)

    #display(df, out)
    out.clear_output()
    with out:
        display(df)


Add.on_click(add_on_click)

items = [Type, Quantity, Strike, Spot, Add]

box_layout = Layout(display='flex',
                    flex_flow='row',
                    align_items='stretch',
                    width='100%')

box_auto = Box(children=items, layout=box_layout)

display_widgets = VBox([box_auto])
display_widgets
```
